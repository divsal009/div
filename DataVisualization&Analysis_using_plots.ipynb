```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#to draw descriptive graph
import seaborn as sns
```

```
boston_df = pd.read_csv('https://raw.githubusercontent.com/ammishra08/MachineLearning/master/Datasets/boston_train.csv')
```

```
boston_df
```

# line plot

```
plt.figure(figsize=(18,9))
plt.plot(boston_df['AGE'])
plt.title('Line Plot')
plt.xlabel('Indexes')
plt.ylabel('AGE')
```

bar **graph**

```
x=[1,3,5,7]
y=[10,30,50,70]
plt.bar(x,y)
```

```
x1=[1,3,5,7]
y1=[10,30,50,70]
x2=[11,13,15,17]
y2=[21,24,26,29]
plt.bar(x1,y1,color='red',label='lab1')
plt.bar(x2,y2,color='green',label='lab2')
plt.legend()
plt.savefig('/content/bar.png')
plt.show()
```

## **histogram**

```
plt.hist(boston_df['MEDV'])
```

```
plt.figure(figsize=(12,8))
plt.hist(boston_df['MEDV'], bins=20)
plt.title('Histogram Plot')
plt.xlabel('Range of MEDV')
plt.ylabel('Frequency')
plt.show()
```

```
iris_df=pd.read_csv('https://raw.githubusercontent.com/ammishra08/MachineLearning/master/Datasets/iris.csv',sep=',')
```

```
iris_df.head()
```

```
iris_df['species'].unique()
```

scatter **plot**(relnship/coreln btn 2 data)

```
plt.scatter(x=iris_df['sepal_length'],y=iris_df['sepal_width'],marker='^')
```

```
plt.scatter(x=iris_df['sepal_length'],y=iris_df['sepal_width'],marker='x')
```

```
plt.scatter(x=iris_df['sepal_length'],y=iris_df['sepal_width'],marker='^',c='green',s=100)
plt.scatter(x=iris_df['sepal_length'],y=iris_df['petal_length'],marker='^',c='red',s=200)
plt.show()
```

```
sns.set_style('darkgrid')
plt.figure(figsize=(13,10))
plt.scatter(x = iris_df['sepal_length'], y = iris_df['sepal_width'], marker = 'x', c = 'red')
plt.scatter(x = iris_df['sepal_length'], y = iris_df['petal_length'], marker = 'x', s = 40)
plt.show()
```

Scatterplot using **seaborn**

```
plt.figure(figsize=(13,10))
sns.scatterplot(x = 'sepal_length', y = 'sepal_width', data=iris_df,hue='sepal_width',size='sepal_width',marker='s')
plt.show()
```

```
plt.figure(figsize=(13,10))
sns.scatterplot(x = 'sepal_length', y = 'sepal_width', data = iris_df, hue='sepal_width', size = 'sepal_width')

plt.figure(figsize=(13,10))
sns.scatterplot(x = 'sepal_length', y = 'sepal_width', data = iris_df, hue='species', marker = 's')
plt.show()
```
